```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv('titanic.csv', index_col = 'PassengerId')
df.head()
```

```python
inputs = df.drop(['Survived', 'Name', 'Cabin', 'Ticket', 'Parch'], axis = 'columns')
inputs.head()
```

```python
target = df.Survived
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
inputs['encoded_Sex'] = encoder.fit_transform(inputs.Sex)
inputs['encoded_Embarked'] = encoder.fit_transform(inputs.Embarked)
inputs.head()
```

```python
processed_inputs = inputs.drop(['Sex', 'Embarked'], axis = 'columns')
processed_inputs.head(7)
```

```python
processed_inputs.head(7)
```

```python
processed_inputs.Age.fillna(processed_inputs.Age.median(), inplace = True)
processed_inputs
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(processed_inputs, target, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
