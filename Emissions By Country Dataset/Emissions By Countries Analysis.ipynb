### Training model to predict emission of CO2 in Pakistan

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
X = np.array(pakistan_emission.Year.tolist())
type(X)
```

```python
y = pakistan_emission['Total']
type(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
model.fit(X_train.reshape(-1,1), y_train)
```

```python
model.coef_
```

```python
model.predict([[2000],[2025],[ 2027]])
```

```python
model_accuracy = model.score(X_test.reshape(-1,1), y_test)
```

```python
print(f'The accuracy of the model is {model_accuracy}%')
```

```python
model.intercept_
```

```python
prediction = model.predict(X.reshape(-1,1))
```

```python
plt.scatter(y, prediction)
```
